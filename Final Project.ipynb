## Preliminary Steps

Import statements needed

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime #needed to convert plaintext 'Date' index into Date type
import pandas_datareader as pdr #data reader to read in stock data from Yahoo Finance
from datetime import datetime
```

We need to create two portfolios, each focusing on a different industry

Putting individual stock tickers and date range into variables

```python
#Home Furnishings portfolio
bbby = 'BBBY' #Bed Bath and Beyond
kirk = 'KIRK' #Kirklands
pir = 'PIR' #Pier 1 Imports Inc.

#Fashion Portfolio
ges = 'GES' #Guess Inc
gps = 'GPS' #The Gap Inc.
jwn = 'JWN' #Nordstrom Inc

begin = datetime(2010, 7, 30)
stop = datetime(2019, 7, 30)
```

Reading stock data into dataframes

```python
# Retrieve data for BBBY from Yahoo Finance
bbby_data = pdr.get_data_yahoo( symbols=bbby, start=begin, end=stop )
bbby_data = bbby_data[['Adj Close']] #updates bbby dataframe to contain only 'Adj Close' column and index 'Date'
bbby_data = bbby_data.rename(columns={'Adj Close': bbby})

# Retrieve data for KIRK from Yahoo Finance
kirk_data = pdr.get_data_yahoo( symbols=kirk, start=begin, end=stop )
kirk_data = kirk_data[['Adj Close']]
kirk_data = kirk_data.rename(columns={'Adj Close': kirk})

# Retrieve data for PIR from Yahoo Finance
pir_data = pdr.get_data_yahoo( symbols=pir, start=begin, end=stop )
pir_data = pir_data[['Adj Close']]
pir_data = pir_data.rename(columns={'Adj Close': pir})

# Retrieve data for GES from Yahoo Finance
ges_data = pdr.get_data_yahoo( symbols=ges, start=begin, end=stop )
ges_data = ges_data[['Adj Close']]
ges_data = ges_data.rename(columns={'Adj Close': ges})

# Retrieve data for GPS from Yahoo Finance
gps_data = pdr.get_data_yahoo( symbols=gps, start=begin, end=stop )
gps_data = gps_data[['Adj Close']]
gps_data = gps_data.rename(columns={'Adj Close': gps})

# Retrieve data for JWN from Yahoo Finance
jwn_data = pdr.get_data_yahoo( symbols=jwn, start=begin, end=stop )
jwn_data = jwn_data[['Adj Close']]
jwn_data = jwn_data.rename(columns={'Adj Close': jwn})
```

```python
# Merge all stocks into one dataframe, then print header only
stock_merged = pd.merge(home_merged,fashion_merged, on='Date')
print(stock_merged.head())

# plot all the raw prices
stock_merged.plot(linewidth=1)
plt.show()
```

```python
#make a copy of stock_merged data
stocks_data_p_change_single_start=stock_merged.copy()

# Convert the raw prices of each stock to percent change from a single starting point for each of the 6 stocks
stocks_data_p_change_single_start[bbby] = (stock_merged[bbby]-stock_merged[bbby][0]) / stock_merged[bbby][0] * 100.0
stocks_data_p_change_single_start[kirk] = (stock_merged[kirk]-stock_merged[kirk][0]) / stock_merged[kirk][0] * 100.0
stocks_data_p_change_single_start[pir] = (stock_merged[pir]-stock_merged[pir][0]) / stock_merged[pir][0] * 100.0
stocks_data_p_change_single_start[ges] = (stock_merged[ges]-stock_merged[ges][0]) / stock_merged[ges][0] * 100.0
stocks_data_p_change_single_start[gps] = (stock_merged[gps]-stock_merged[gps][0]) / stock_merged[gps][0] * 100.0
stocks_data_p_change_single_start[jwn] = (stock_merged[jwn]-stock_merged[jwn][0]) / stock_merged[jwn][0] * 100.0

# print header
print(stocks_data_p_change_single_start.head())

# plot converted raw price to percent change
stocks_data_p_change_single_start.plot(linewidth=0.6)
plt.show()
```
